```python
import yfinance as yf
import pandas as pd
import statsmodels.api as sm
import matplotlib.pyplot as plt
from scipy import stats

# Helper function to remove timezone from datetime objects
def remove_timezone(df):
    df.index = df.index.tz_localize(None)
    return df

# List of stocks
stocks = ['WIPRO.NS', 'TATASTEEL.NS', 'TATAMOTORS.NS', 'TATAMTRDVR.NS',
          'TCS.NS', 'SUNPHARMA.NS', 'SBIN.NS', 'POWERGRID.NS',
          'ONGC.NS', 'NTPC.NS']

# Nifty 50 as the benchmark index
nifty = "^NSEI"

# Event dates (Indian elections in 2014 and 2019)
event_dates = ['2014-05-16', '2019-05-23']

# Define event window and estimation window
event_window = 10  # 10 days before and after the event
estimation_window = 100  # 100 days before the event

# Download stock and market data
start_date = '2013-01-01'
end_date = '2020-12-31'

# Fetch stock and Nifty data from Yahoo Finance
stock_data = yf.download(stocks, start=start_date, end=end_date)['Adj Close']
nifty_data = yf.download(nifty, start=start_date, end=end_date)['Adj Close']

# Remove time zone information from stock and Nifty data
stock_data = remove_timezone(stock_data)
nifty_data = remove_timezone(nifty_data)

# Calculate daily returns for stocks and Nifty
stock_returns = stock_data.pct_change().dropna()
nifty_returns = nifty_data.pct_change().dropna()

# Function to estimate market model and calculate abnormal returns
def calculate_abnormal_returns(stock_returns, market_returns, event_date):
    # Convert event_date to timezone-naive if necessary
    event_date = pd.to_datetime(event_date).tz_localize(None)

    # Define estimation and event window
    estimation_start = pd.to_datetime(event_date) - pd.DateOffset(days=estimation_window + event_window)
    estimation_end = pd.to_datetime(event_date) - pd.DateOffset(days=1)
    event_start = pd.to_datetime(event_date) - pd.DateOffset(days=event_window)
    event_end = pd.to_datetime(event_date) + pd.DateOffset(days=event_window)

    # Align the stock and market returns to ensure matching indices
    stock_returns_aligned, market_returns_aligned = stock_returns.align(market_returns, join='inner')

    # Subset the estimation window data
    estimation_returns = stock_returns_aligned[estimation_start:estimation_end]
    event_returns = stock_returns_aligned[event_start:event_end]
    market_estimation_returns = market_returns_aligned[estimation_start:estimation_end]
    market_event_returns = market_returns_aligned[event_start:event_end]

    # Estimate the market model (OLS regression)
    X = sm.add_constant(market_estimation_returns)
    model = sm.OLS(estimation_returns, X).fit()
    alpha, beta = model.params

    # Calculate predicted returns during the event window
    predicted_returns = alpha + beta * market_event_returns

    # Calculate abnormal returns
    abnormal_returns = event_returns - predicted_returns

    return abnormal_returns, model

# Function to calculate cumulative abnormal returns and t-test
def event_study(stock, stock_returns, nifty_returns, event_dates):
    for event_date in event_dates:
        abnormal_returns, model = calculate_abnormal_returns(stock_returns[stock], nifty_returns, event_date)
        CAR = abnormal_returns.cumsum()

        # Perform t-test to check for significance of abnormal returns
        t_stat, p_value = stats.ttest_1samp(abnormal_returns, 0)

        print(f'\n{stock} - Event Date: {event_date}')
        print(f'T-statistic: {t_stat}, P-value: {p_value}')
        print(f'Cumulative Abnormal Return (CAR): {CAR[-1]}\n')

        # Plotting abnormal returns and CAR
        plt.figure(figsize=(10, 6))
        plt.plot(abnormal_returns.index, abnormal_returns, label='Abnormal Returns', marker='o')
        plt.plot(CAR.index, CAR, label='Cumulative Abnormal Returns (CAR)', marker='o')
        plt.axvline(x=pd.to_datetime(event_date), color='r', linestyle='--', label='Event Date')
        plt.title(f'{stock} Abnormal Returns & CAR around Event ({event_date})')
        plt.xlabel('Date')
        plt.ylabel('Returns')
        plt.legend()
        plt.show()

# Run event study for each stock
for stock in stocks:
    event_study(stock, stock_returns, nifty_returns, event_dates)


```
